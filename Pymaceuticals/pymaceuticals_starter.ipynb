## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)


# Combine the data into a single dataset

combined_data_df = pd.merge(mouse_metadata,study_results,on="Mouse ID",how="outer")
combined_data_df.head()
#combined_data_df.shape
```

## Summary statistics

```python
combined_group = combined_data_df.groupby("Drug Regimen")
combined_group["Drug Regimen"].count()

#combined_data_df.describe()

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


from tabulate import tabulate

drug_regimens = combined_data_df["Drug Regimen"].unique()
index=0
summ_stats = []

for i in drug_regimens:
    select_temp = combined_data_df[combined_data_df["Drug Regimen"] ==i]
    avg = select_temp["Tumor Volume (mm3)"].mean()
    med = select_temp["Tumor Volume (mm3)"].median()
    var = select_temp["Tumor Volume (mm3)"].var()
    std = select_temp["Tumor Volume (mm3)"].mean()
    sem = select_temp["Tumor Volume (mm3)"].sem()
    tot_count = select_temp["Tumor Volume (mm3)"].count()
    summ_stats.append([i,tot_count,round(avg,2),round(med,2),round(var,2),round(std,2),round(sem,2)])
    index+= 1
    
    table = tabulate (summ_stats,headers=['Drug Regimen','count','mean','median',
                                          'variance','standard deviation','SEM'],tablefmt='pretty')

print("Summary Statistics Table")

print(table)
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers.


drug_regimens = ['Capomulin','Ramicane', 'Infubinol', 'Ceftamin']

for i in drug_regimens:
    df_drug = combined_data_df[combined_data_df["Drug Regimen"] == i]
    df_tumor = df_drug['Tumor Volume (mm3)']
    quartiles = df_tumor.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = abs(upperq-lowerq)
    print('----------------------' + i + '----------------------')
    print(f"The lower quartile is: {round(lowerq,2)}")
    print(f"The upper quartile is: {round(upperq,2)}")
    print(f"The interquartile range is: {round(iqr,2)}")
  
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {round(lower_bound,2)} could be outliers.")
    print(f"Values above {round(upper_bound,2)} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest


fig = plt.figure(figsize = (7,4))
final_TV = []

for i in drug_regimens:
    df_drug = combined_data_df[combined_data_df["Drug Regimen"] == i]
    df_tumor = df_drug['Tumor Volume (mm3)']
    final_TV.append(df_tumor)
diamond = dict(markerfacecolor='g', marker='D')
plt.boxplot(final_TV,whis=1.5,labels=drug_regimens,flierprops=diamond)

plt.ylabel('Final Tumor Volume  (mm3)')
plt.grid(linestyle=':')
plt.tight_layout()
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin



# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

merged_data = pd.merge(study_results, mouse_metadata, on='Mouse ID',how='outer')
merged_data

#Create df for mouse u364 
cap_df = merged_data.loc[(merged_data["Drug Regimen"] == 'Capomulin') & (merged_data['Mouse ID'] == 'u364')]
[["Tumor Volume (mm3)", "Timepoint"]]

#plot the change over time for mouse u364
x_axis = cap_df["Timepoint"]
y_axis = cap_df["Tumor Volume (mm3)"]
finalVolume, = plt.plot(x_axis, y_axis)

plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Volume over Time for Capomulin Mouse u364')
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen


cap_scatter = merged_data.loc[merged_data["Drug Regimen"] == 'Capomulin']

#find the average for weight and tumor volume
x_axis_scatter = cap_scatter.groupby(cap_scatter["Mouse ID"])["Weight (g)"].mean()
y_axis_scatter = cap_scatter.groupby(cap_scatter["Mouse ID"])["Tumor Volume (mm3)"].mean()

#plot
plt.scatter(x_axis_scatter, y_axis_scatter, marker="o", facecolors="red")
plt.xlabel("Weight of Mouse")
plt.ylabel("Tumor Volume")
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
from scipy import stats
x_values = cap_scatter.groupby(cap_scatter["Mouse ID"])["Weight (g)"].mean()
y_values = cap_scatter.groupby(cap_scatter["Mouse ID"])["Tumor Volume (mm3)"].mean()


m_slope, m_int, m_r, m_p, m_std_err = stats.linregress(x_values, y_values)
m_fit = m_slope * x_values + m_int
plt.scatter(x_values, y_values)
plt.plot(x_values,m_fit,"--")

plt.title('Mouse Weight versus Average Tumor Volume, Capomulin regimen')
plt.xlabel("Mouse Weight")
plt.ylabel("Average Tumor Volume (mm3)")

line_eq = "y = " + str(round(m_slope,2)) + "x + " + str(round(m_int,2))
plt.annotate(line_eq,(20,40),fontsize=15,color="red")

cc = stats.pearsonr(x_values, y_values)
cc_annot = 'correlation coefficient = '+ str(round(cc[0],2))
plt.annotate(cc_annot,(15,44),fontsize=15,color="red")
```
